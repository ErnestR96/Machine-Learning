Sistema de recomendación basado en el contenido

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.neighbors import NearestNeighbors
```

```python
imdb = pd.read_csv('imdb_top_1000.csv')
imdb.head()
```

```python
imdb.info()
```

```python
# Trabajaremos unicamente con las columnas Series_title, Genre, IMDB_Rating, Director, Star1, Star2
columnas = ['Series_Title','Genre','IMDB_Rating','Director','Star1','Star2']
imdb_recomendacion = imdb[columnas]
imdb_recomendacion.head()
```

```python
#Aplicamos che a las variables categóricas
imdb_recomendacion_t = pd.get_dummies(
    imdb_recomendacion,
    columns = ['Genre', 'Director', 'Star1', 'Star2']
    )
imdb_recomendacion_t
```

```python
imdb_recomendacion_t.shape
```

```python
#Creamos el modelo
knn = NearestNeighbors(n_neighbors= 10)
knn
```

```python
#Ajustamos el modelo de recomendación
recomendacion = knn.fit(imdb_recomendacion_t.drop(columns = ['Series_Title']))
```

```python
#Hacemos las predicciones
distancias, indices = recomendacion.kneighbors(imdb_recomendacion_t.drop(columns=['Series_Title']))
```

```python
indices[1]
distancias[1]
```

```python
indice = imdb_recomendacion_t[imdb_recomendacion_t['Series_Title'] == 'The Dark Knight'].index[0]
Pelicula = imdb_recomendacion_t.loc[indices[indice], 'Series_Title']
Calificacion = imdb_recomendacion_t.loc[indices[indice], 'IMDB_Rating']
Calificacion
```

```python
#Creamos una función para hacer las recomendaciones
def recomendacion_pelicula(pelicula):
    #Buscamos el indice de la pelicula
    indice = imdb_recomendacion_t[imdb_recomendacion_t['Series_Title'] == pelicula].index[0]
    #Creamos un dataframe con las peliculas mas cercanas
    resultado = pd.DataFrame(
        {
            'Pelicula' : imdb_recomendacion.loc[indices[indice],'Series_Title'],
            'Calificacion' : imdb_recomendacion.loc[indices[indice], 'IMDB_Rating']
         }   
        )
    return resultado

recomendacion_pelicula('The Dark Knight')
```
