Perceptron

Modelo que permite calcular la probabilidad de obtener matrícula de honor al final del bachillerato en función de la nota que se ha obtenido en matemáticas.
0 = no tiene matrícula
1 = si tiene matrícula

```python
#Tratamiento de lo datos
import numpy as np
import pandas as pd

#Gráficos
import matplotlib.pyplot as plt
```

```python
#Preprocesado y modelado
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score
import warnings
warnings.filterwarnings('ignore')
```

```python
#Datos

n_samples = 500
X, y = make_blobs(n_samples=n_samples,
                            centers=([1.1,3],[4.5,6.9]),
                            cluster_std= 1.3,
                            random_state=0)

colours = ('green','orange')
fig, ax = plt.subplots()
plt.grid()
for n_class in range(2):
    ax.scatter(X[y==n_class][:,0],
               X[y==n_class][:,1],
               c = colours[n_class],
               s = 50,
               label = str(n_class))
    
```

```python
#División de los datos en train y test
datasets = train_test_split(X,
                            y,
                            test_size=0.2)
X_train, X_test, y_train, y_test = train_test_split(
                            X,
                            y,
                            train_size= 0.8,
                            random_state= 1234,
                            shuffle = True
                            )
```

```python
#creación del modelo
modelo = Perceptron(random_state=42)
modelo
```

```python
#Entrenamiento del modelo
modelo.fit(X_train, y_train)
```

```python
#Error de test del modelo
predicciones_train = modelo.predict(X = X_train)
predicciones_test = modelo.predict(X = X_test)
```

```python
#Visualizar predicciones
print('Datos esperados - Predicciones')
print('---------------------')
print(np.column_stack((y_test, predicciones_test)))
```

```python
train_score = accuracy_score(predicciones_train, y_train)
print(f'El accuracy de train es: {100 * train_score} %')
test_score = accuracy_score(predicciones_test, y_test)
print(f'El accuracy de test es: {100 * test_score} %')
```
